```python
#checking the dependency of different regions in promotion
data=pd.crosstab(train['region'],train['is_promoted'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar',stacked=True,figsize=(20,8),color=['cyan','pink'])
plt.title('dependency of regions determining promotion of employee')
plt.xlabel("different regions of the company",fontsize=20)
plt.legend()
plt.show()
```

```python
#checking the dependency of awards won on promotion
data=pd.crosstab(train['awards_won?'],train['is_promoted'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar',stacked=True,figsize=(20,8),color=['cyan','pink'])
plt.title('dependency of awards in determining promotion of employee',fontsize=20)
plt.xlabel("awards won or not",fontsize=20)
plt.legend()
plt.show()
```

```python
#checking the dependency of age factor in promotion
data=pd.crosstab(train['age'],train['is_promoted'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar',stacked=True,figsize=(20,8),color=['cyan','pink'])
plt.title('dependency of age factor in determining promotion of employee',fontsize=20)
plt.xlabel("age factor",fontsize=20)
plt.legend()
plt.show()
```

```python
#checking which department got the greatest number of promotion
data=pd.crosstab(train['department'],train['is_promoted'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar',stacked=True,figsize=(20,8),color=['cyan','pink'])
plt.title('department got maximun promotion',fontsize=20)
plt.xlabel("department who promoted max",fontsize=20)
plt.legend()
plt.show()
```

```python
#checking dependency of gender over promotion
data=pd.crosstab(train['gender'],train['is_promoted'])
data.div(data.sum(1).astype('float'),axis=0).plot(kind='bar',stacked=True,figsize=(20,8),color=['cyan','pink'])
plt.title('dependency of gender over promotion',fontsize=20)
plt.xlabel("gender and promotion",fontsize=20)
plt.legend()
plt.show()
```
